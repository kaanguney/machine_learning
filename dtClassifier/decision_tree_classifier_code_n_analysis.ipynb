```python
import numpy as np
import sys # for interpreter-function-library accesses 
from scipy import stats
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# necessary libraries imported 
```

```python
# import the dataset from the archive
```

```python
bank_data = pd.read_csv("bank-full.csv", sep = ";")
bank_data.head() 
```

```python
# y is our target variable, change the name for ease
bank_data = bank_data.rename(columns = {"y" : "Target"})
bank_data.head()
```

```python
# split the dataset into two groups dependent(target) variable and independent variables(features)
```

```python
feature_cols = ["age", "job", "marital", "education", "default", "balance", "housing", "loan", "contact", "day", "month", "duration", "campaign", "pdays", "previous", "poutcome"] 
```

```python
X = bank_data[feature_cols] # features 
y = bank_data["Target"] # target variable
```

```python
from sklearn.tree import DecisionTreeClassifier 
from sklearn.model_selection import train_test_split
from sklearn import metrics # for accuracy
```

```python
# split the data into training and test set 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 1)
```

```python
X_train.dtypes # have to convert object datatypes using labelencoder before using fit()
```

```python
from sklearn import preprocessing # to use for labelencoder(), preprocess both sets 

def convert(data): 
    le = preprocessing.LabelEncoder()
    for i in data:
        if(data[i].dtype == object):
            data[i] = le.fit_transform(data[i])
        else:
            pass
    return data 
X_train = convert(X_train)
X_test = convert(X_test)
```

```python
tree = DecisionTreeClassifier() # Decision Tree Object, max_depth created max by default for first depth tryout 

tree = tree.fit(X_train, y_train) # Train decision tree classifier 

y_pred = tree.predict(X_test) # predict the outcome for test set
```

```python
# calculate model accuracy, how often is classifier accurate?

print("Accuracy of the model is ", metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, y_pred)) # just additional info, not really needed 
```

```python
# f1 score is almost the same as accuracy, meaning we have an even distribution 
# if that wasn't the case; f1-score would provide better results compared to accuracy metric
```

### - We can also specify the parameters for our decision tree model such as max_depth manually...

```python
# reconstructing the tree to measure if max_depth specification changes the accuracy
# use a for loop to try different depths
```

```python
for i in range(1,11,1):
    clf = DecisionTreeClassifier(max_depth = i)
    clf = clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    print("Accuracy of the model with max_depth = ",i," ",float("%.5f" % metrics.accuracy_score(y_test, y_pred)))
```

### The results show that; 
************************************

    - The most accurate model turned out to be in max_depth = 9 (out of 10 different tryouts) - %90.065 success rate.
    
    - Models with max_depth = 6, 7, 8 and 9 have very close accuracies ~ %90.
